### CONCATENATE

```python
import pandas as pd
```

```python
india_weather  = pd.DataFrame({
                "city" : ['Mumbai','Delhi','Jamshedpur'],
                 "temp" : [35,40,36],
                 "humidity" : [70,60,65]
             })
india_weather
```

```python
us_weather = pd.DataFrame({
    "city" : ['New York', 'California', 'Arizona'] ,
    "temp" : [27,24,36],
    "humidity" : [65,55,50]
})

us_weather
```

```python
df = pd.concat([india_weather,us_weather])
df
```

```python
df1 = pd.concat([india_weather,us_weather], ignore_index=True)
df1
```

```python
df2=pd.concat([india_weather,us_weather],keys=['INDIA','US'])
df2
```

```python
df2.loc['INDIA']
```

```python
df2.loc['US']
```

### MERGE

```python
india_temp = pd.DataFrame({'city':['mumbai','delhi','jamshedpur'],
                        'temp':[35,45,37]
                        })
india_temp
```

```python
india_humidity = pd.DataFrame({'city':['mumbai','delhi','jamshedpur'],
                              'humidity':[60,50,55]
                              })
india_humidity
```

```python
df_merged = pd.merge(india_temp,india_humidity, on ='city')
df_merged
```

Below example shows even if the order of cities is different in the two dataframes it still joins them correctly as it uses city on which join is performed

```python
us_temp = pd.DataFrame({'city':['NewYork','California','Arizona'],
                        'temp':[24,23,40]
                        })
us_temp
```

```python
us_humidity = pd.DataFrame({'city':['NewYork','Arizona','California'],
                              'humidity':[60,55,50]
                              })
us_humidity
```

```python
df_merged_us = pd.merge(us_temp, us_humidity, on ='city')
df_merged_us
```

Below example shows how the resultant dataframe will look if both dataframes have different no of rows for the columns on which join is performed

```python
us_temp = pd.DataFrame({'city':['NewYork','California','Arizona','Baltimore'],
                        'temp':[24,23,40,33]
                        })
us_temp
```

```python
us_humidity = pd.DataFrame({'city':['NewYork','Arizona','Washington DC','New jersy'],
                              'humidity':[60,55,42,21]
                              })
us_humidity
```

```python
df_merged_us = pd.merge(us_temp, us_humidity, on ='city') # By default the join is inner(intesection) join
df_merged_us
```

```python
df_merged_us = pd.merge(us_temp, us_humidity, on ='city',how='outer') # Here we are specifying the join to be outer
df_merged_us
```

```python
df_merged_us = pd.merge(us_temp, us_humidity, on ='city',how='left')

df_merged_us
```

```python
df_merged_us = pd.merge(us_temp, us_humidity, on ='city',how='right', indicator=True)

df_merged_us
```

```python
india_weather  = pd.DataFrame({
                "city" : ['Mumbai','Delhi','Jamshedpur','Bangalore'],
                 "temp" : [35,40,36,29],
                 "humidity" : [70,60,65,72]
             })
india_weather
```

```python
india_weather_1  = pd.DataFrame({
                "city" : ['Delhi','Jamshedpur'],
                 "temp" : [42,39],
                 "humidity" : [61,69]
             })
india_weather_1
```

```python
#Now if we want to merge these two df then it will automatically add the column suffixes to distinguish the df from where it is coming
```

```python
df = pd.merge(india_weather,india_weather_1, on='city')
df
```

```python
# But if we want to specify the columns name then we can provide the suffixes
df = pd.merge(india_weather,india_weather_1, on='city',suffixes=['_left','_right'])
df
```
