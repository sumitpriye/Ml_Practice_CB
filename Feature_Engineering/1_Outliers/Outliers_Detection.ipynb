```python
min_threshold, max_threshold = df_bhp.price_per_sqft.quantile([0.001, 0.999])
min_threshold, max_threshold
```

So, based upon the above min & max threshold values we will consider that houses having 'price_per_sqft' less than 1366.184 and greater than 50959 as outliers

```python
df_bhp[df_bhp.price_per_sqft<min_threshold]
```

```python
df_bhp[df_bhp.price_per_sqft>max_threshold]
```

```python
df_bhp2 = df_bhp[(df_bhp.price_per_sqft > min_threshold) & (df_bhp.price_per_sqft < max_threshold)]
df_bhp2
```

```python
df_airbnb.reviews_per_month.mean()
```

```python
df_airbnb.isna().sum()
```

```python
min_threshold, max_threshold = df_airbnb['price'].quantile([0.01,0.99])
min_threshold, max_threshold
```

```python
df_airbnb[df_airbnb.price<min_threshold]
```

```python
df_airbnb[df_airbnb.price>max_threshold]
```

```python
df_airbnb = df_airbnb[(df_airbnb.price>min_threshold) & (df_airbnb.price<max_threshold)]
```
