```python
# Basic libraray imports

import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
```

```python
# Reading the Wine dataset from sklearn datasets

from sklearn.datasets import load_wine
wine = load_wine()
dir(wine)
```

```python
wine.target
```

```python
df = pd.DataFrame(data=wine.data,columns=wine.feature_names)
df.head()
```

```python
df['Class'] = wine.target
```

```python
df[df['Class']==2].head()
```

```python
df['Class_Name']=df.Class.apply(lambda x : wine.target_names[x])
df.head()
```

```python
df.describe()
```

```python
df.isna().any()
```

```python
df.info()
```

```python
df
```

```python
target=df['Class']
target
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
scaler.fit(df.drop(['Class_Name','Class'],axis=1))
```

```python
df=scaler.transform(df.drop(['Class_Name','Class'],axis=1))
```

```python
df=pd.DataFrame(data=df)
```
