```python
df_with_dummies=pd.get_dummies(df,drop_first=True)
```

```python
X=df_with_dummies.drop('HeartDisease',axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Scaling should be done post tarin test split otherwise data leakage may take place. The same scaler which is sued for fir the train data should be used for transforming the train and test data

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_X_train=scaler.fit_transform(X_train)
scaled_X_train
```

```python
scaled_X_test=scaler.transform(X_test)
scaled_X_test
```

```python
# Using SVM as a standalone alogorithm for building the model 

from sklearn.svm import SVC
clf_svm = SVC(C=1.0,
    kernel='rbf',
    degree=3,
    gamma='auto',
    random_state=2)
```

```python
clf_svm.fit(scaled_X_train,y_train)
```

```python
clf_svm.score(scaled_X_train,y_train)
```

```python
clf_svm.score(scaled_X_test,y_test)   # Standalone SVM Model score on scaled test data set
```

```python
# Using Grid search to find the best parameters(Hyper parameter tuning)

from sklearn.model_selection import GridSearchCV
grid_clf = GridSearchCV(SVC(gamma='auto'),{
           'C':[1,5,10],
           'kernel':['rbf','linear']          
},cv=5, return_train_score=False, verbose=10)

grid_clf.fit(scaled_X_train,y_train)
```

```python
grid_clf.cv_results_
```

```python
# importing above results into a dataframe
df_gridsearch=pd.DataFrame(grid_clf.cv_results_)
df_gridsearch
```

```python
df_gridsearch[['param_C','param_kernel','mean_test_score']] # So the best training score is for kernel=rbf and C=1 
```

```python
grid_clf.score(scaled_X_test,y_test)  # Score for Scaled test data set using the GRID SEARCH CV 
```

Now creating a bagging model using SVM

```python
from sklearn.ensemble import BaggingClassifier
bag_model = BaggingClassifier(
    base_estimator=SVC(C=1.0,kernel='rbf',gamma='auto'),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=0)
```

```python
bag_model.fit(scaled_X_train,y_train)
```

```python
bag_model.oob_score_
```

```python
bag_model.score(scaled_X_test,y_test)  # Scaled test data score for bagging model using SVC 
```

```python
bag_model.score(scaled_X_train,y_train)
```

```python
# Using grid search cv along with bagging
from sklearn.ensemble import BaggingClassifier
bag_model = BaggingClassifier(
    base_estimator=SVC(gamma='auto'),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=0)

bag_grid_clf = GridSearchCV(bag_model,{
    'base_estimator__C':[1,5,10],'base_estimator__kernel':['rbf','linear']},
     cv=5, return_train_score=False, verbose=10)

bag_grid_clf.fit(scaled_X_train,y_train)
```

```python
bag_grid_clf.cv_results_
```

```python
# importing above results into a dataframe
df_bag_gridsearch=pd.DataFrame(bag_grid_clf.cv_results_)
df_bag_gridsearch
```

```python
df_bag_gridsearch[['params','mean_test_score']]
```

```python
bag_grid_clf.score(scaled_X_test,y_test) # Scaled test data score using Grid Search CV along with bagging  
```
