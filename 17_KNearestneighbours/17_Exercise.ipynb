```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
dir(digits)
```

```python
df = pd.DataFrame(data=digits.data,columns=digits.feature_names)
df
```

```python
df['target'] = digits.target
```

```python
df
```

```python
input=df.drop('target',axis=1)
```

```python
input
```

```python
target=df['target']
```

```python
target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(input,target,test_size=0.2,random_state=2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=9)
```

```python
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

```python
knn.score(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix
y_pred=knn.predict(X_test)
cm=confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
import seaborn as sns
plt.figure(figsize=(10,8))
sns.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test,y_pred)
```

```python
report
```

```python
print(report)
```

```python
# Using Grid Search

from sklearn.model_selection import GridSearchCV

clf = GridSearchCV(KNeighborsClassifier(),{
    'n_neighbors':[1,2,3,5,7,10,20,50]
},cv=5, return_train_score=False,verbose=10)
#clf.fit(input,target)
clf.fit(X_train,y_train)
clf.cv_results_
```

```python
# importing above results into a dataframe
df_gridsearch=pd.DataFrame(clf.cv_results_)
df_gridsearch
```
